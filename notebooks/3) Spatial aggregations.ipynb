```python
traffic.rename(columns={'latitud': 'LATITUD', "longitud":"LONGITUD"}, inplace=True)
traffic.info()
```

```python
# Convert LATITUD and LONGITUD columns to numeric
def process_coordinates(df):
    df['LATITUD'] = pd.to_numeric(df['LATITUD'], errors='coerce')
    df['LONGITUD'] = pd.to_numeric(df['LONGITUD'], errors='coerce')
    # Drop rows with missing LATITUD or LONGITUD
    df.dropna(subset=['LATITUD', 'LONGITUD'], inplace=True)
    return df

# Process the dataframes
traffic = process_coordinates(traffic)
meteo = process_coordinates(meteo)
airqual = process_coordinates(airqual)

# Create GeoDataFrame
def create_geodataframe(df):
    gdf = gpd.GeoDataFrame(
        df, geometry=gpd.points_from_xy(df.LONGITUD, df.LATITUD))
    return gdf

traffic_gdf = create_geodataframe(traffic)
meteo_gdf = create_geodataframe(meteo)
airqual_gdf = create_geodataframe(airqual)
```
